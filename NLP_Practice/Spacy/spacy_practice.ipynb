```python
import spacy
```

```python
nlp=spacy.load("en_core_web_lg")
```

```python
text1 = "Dr. A.P.J. Abdul Kalam was born in Rameswaram, Tamil Nadu, India. He later became the President of India and inspired millions of students."

text2 = "Today, Dhuru is learning Machine Learning and Natural Language Processing at Nyeras in Bangalore."

text3 = "He created an NLP project for Twitter sentiment analysis using Python, Flask, scikit-learn, and spaCy."

text4 = "On 5 May 2026, he sent an email to support@nyeras.com and visited https://www.nyeras.com for project details."

text5 = "The course fee was ₹15,000, but the discounted price was $120."

text6 = "I love this product! The user-friendly design, fast performance, and clean interface make it amazing."

text7 = "However, the app sometimes crashes when I upload a large CSV file."

text8 = "Apple Inc. was founded by Steve Jobs, Steve Wozniak, and Ronald Wayne in California."

text9 = "Google announced a new AI model in Tokyo, Japan, on Monday."

text10 = "The student who studied for 8 hours scored 92 marks in the exam."

text11 = "The student who studied for 2 hours scored only 45 marks."

text12 = "Book a flight from Chennai to Tokyo tomorrow morning."

text13 = "Please remind me to submit the machine learning assignment before 10:30 PM."

text14 = "Contact me at dhuru.data@gmail.com or call +91-9876543210."

text15 = "The invoice number is INV-2026-ML-045, and the order ID is ORD#A12345."

text16 = "I can't believe this model predicted the result correctly!"

text17 = "This is awesome, but the preprocessing step took too much time."

text18 = "Mr. Raj bought 3 laptops for ₹1,80,000 from Amazon India."

text19 = "The delivery address was No. 24, Anna Nagar, Chennai, Tamil Nadu."

text20 = "Although the dataset was small, the model performed well because the features were meaningful."

text21 = "If the text is cleaned properly, the machine learning model can understand the pattern better."

text22 = "New York is bigger than Boston, but Tokyo is one of the most advanced cities in the world."

text23 = "Dhuruvan Shanker R is planning to work in a Japanese IT company after gaining experience in India."
```

```python
from spacy.symbols import ORTH

nlp.tokenizer.add_special_case("Dr.",[
    {ORTH:"Dr."}
])

nlp.tokenizer.add_special_case("A.P.J.",[
    {ORTH:"A.P.J."}
])
```

```python
tokens1=[]
doc1=nlp(text1)
for token in doc1:
    tokens1.append(token)
tokens1
```

```python
tokens1=[]
doc1=nlp(text1)
for token in doc1:
    if not token.is_punct:
        tokens1.append(token)
tokens1
```

```python
nlp.add_pipe("sentencizer")
```

```python
tokens1=[]
doc1=nlp(text1)
for token in doc1.sents:
    tokens1.append(token)
tokens1
```

```python
doc2 = nlp(text2)

for token in doc2:
    if not token.is_stop and not token.is_punct and token.is_alpha:
        print(
            "Token:", token.text,"\n",
            "| Parent/Head:", token.head.text,"\n",
            "| Children:", [child.text for child in token.children],"\n",
            "| Dependency:", token.dep_,"\n",
            "| Meaning:", spacy.explain(token.dep_),"\n",
        )
```

```python
doc4=nlp(text4)

for token in doc4:
    if token.like_num:
        print("Like num token : ",token)
    if token.like_email: 
        print("Like email token : ",token)
    if token.like_url: 
        print("Like url token : ",token)
```

```python
doc5=nlp(text5)

for token in doc5:
    if token.like_num:
        print("Like num token : ",token)
    if token.is_currency: 
        print("currency token : ",token)
```

```python
doc18=nlp(text18)
from spacy import displacy
displacy.render(doc18,style='dep')
```

```python
doc8=nlp(text8)
from spacy import displacy
displacy.render(doc8,style='ent')
```

```python
doc15=nlp(text15)
for token in doc15:
    print("Original text :",token.text,"|","Token's format :",token.shape_,"\n")
```

```python
doc12=nlp(text12)
for token in doc12:
    print("Token's original form :",token.text,"\n","Token's grammar role :",token.pos_,"\n","Explanation of token's grammar role :",spacy.explain(token.pos_),"\n","Token's Detailed grammar role :",token.tag_,"\n","Detailed explanation of token's grammar role :",spacy.explain(token.tag_),"\n")
```

```python
doc9=nlp(text9)
from spacy import displacy
displacy.render(doc9,style='ent')
```

```python
doc23=nlp(text23)
for token in doc23:
    print("Original text :",token.text,"\n","Base form :",token.lemma_,"\n")
```

```python
doc10=nlp(text10)
for token in doc10:
    print("Original text :",token.text,"\n","Base form :",token.lemma_,"\n")
```
