```python
corpus="I am Dhuru, and I am currently learning Machine Learning and Natural Language Processing. I love building projects, and I am working on improving my skills every day. Running, coding, and analyzing data are my favorite activities!"
print(corpus)
```

```python
from nltk.tokenize import sent_tokenize
sentences=sent_tokenize(corpus)
print(sentences)
```

```python
for i in range(len(sentences)):
    print(f'Sentence no : {i+1} --> {sentences[i]}')
```

```python
from nltk.tokenize import word_tokenize
words=[word_tokenize(sentences[i]) for i in range(len(sentences))]
print(words)
print()
for i in range(len(words)):
    print(f'No:of words in the sentence {i+1} --> {len(words[i])}')
```

```python
from nltk.corpus import stopwords
for i in range(len(words)):
    words[i]=[word for word in words[i] if word.lower() not in set(stopwords.words('english'))]
    print(f'Removal of stopwords for sentence {i+1} --> {words[i]}')
```

```python
for i in range(len(words)):
    words[i]=[word.lower() for word in words[i] if word.isalpha()]
    print(f'Removal of punctuation marks("." , ",") for sentence {i+1} --> {words[i]}')
```

```python
from nltk import pos_tag
tags=pos_tag(words[1])
print(tags)
print(type(tags))
print(type(tags[1]))
print(tags[1])
print(tags[1][0])
```

```python
for word,tag in tags:
    print(word,tag)
```

```python
from nltk.stem import SnowballStemmer
stemming=SnowballStemmer('english')

stemming_words=[]
for i in range(len(words)):
    stemmed_words=[stemming.stem(word) for word in words[i]]
    stemming_words.append(stemmed_words)
    print(f'Stemmed words for sentence {i+1} --> {stemming_words[i]}')
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizier=WordNetLemmatizer()

def wordnet_pos(tag):
        if tag.startswith("N"):
            return 'n'
        elif tag.startswith("V"):
            return 'v'
        elif tag.startswith("J"):
            return 'a'
        elif tag.startswith("R"):
            return 'r'
        else:
            return 'n'

for i in range(len(words)):
    word=words[i]
    words_tags=pos_tag(words[i])
    tag=[tags for word,tags in words_tags]
    temp=[]
    for j in range(len(tag)):
        word_pos=wordnet_pos(tag[j])
        temp.append(lemmatizier.lemmatize(word[j],pos=word_pos))
    words[i]=temp 

for i in range(len(words)):
    print(f'Sentence {i+1} words after lemmatization --> {words[i]}')
```

```python
for i in range(len(sentences)):
    print(f'Old sentence {i+1} --> {sentences[i]}')
    sentences[i]=" ".join(words[i])
    print(f'New sentence {i+1} --> {sentences[i]}')
    print()
```
