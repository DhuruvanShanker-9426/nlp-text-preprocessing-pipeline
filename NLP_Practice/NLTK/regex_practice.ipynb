```python
import re
```

```python
digits_pattern=re.compile(r"\d+")
```

### **Extract all numbers using regex**

```python
text = "My marks are 85 and 92"
numbers=digits_pattern.findall(text)
print(f'The numbers from the text : {numbers}')
```

### **Check if string starts with "hello"**

```python
text = "hello world"
ans=re.match(r"hello",text)
res="'hello' word is found at the beginning of the text" if ans else  "'hello is not at the beginnig of the text'"
print(res)
#print(ans)
```

### **Extract only the first number**

```python
text = "Age: 21, Marks: 95"
first_num=digits_pattern.search(text)
print(f'The first number in the text is : {first_num.group()}')
```

### **Find all vowels**

```python
text = "machine learning"
vowels_list=re.findall(r"[aeiou]",text)
print(f'All vowels from the text : {vowels_list}')
```

### **Extract only words (ignore numbers)**

```python
text = "Hello123 World456"
words=re.findall(r"[a-zA-Z]+",text)
print(f'Only words in the text : {words}')
```

### **Remove everything except letters and numbers**

```python
text = "Hello!!! @ML #2026"
cleaned_text=re.sub(r"[^a-zA-Z0-9\s]+","",text)
print(f'Removed the special characters from the text : {cleaned_text}')
```

### **Extract only numbers with exactly 3 digits**

```python
text = "123 45 6789 456"
three_digits=re.findall(r"\b\d{3}\b",text)
print(f'Three digits in the text : {three_digits}')
```

### **Match correct spellings using regex**

```python
text = ["color", "colour", "colouur"]
correct_words=[word for word in text if re.fullmatch(r"colou?r", word)]
print(f'Correct spelling words : {correct_words}')
```

### **Convert multiple spaces → single space**

```python
text = "I   love    ML"
extra_to_singlespace=re.sub(r"\s+"," ",text)
print(f'Removed the extra spaces in the text : {extra_to_singlespace}')
```

### **Split using , ; |**

```python
text = "apple,banana;grape|orange"
splitted_words=re.split(r"[,;|]",text)
print(f'Text converted into list of words : {splitted_words}')
```

### **Extract email**

```python
texts = [
    "Hey, contact me at dhuru123@gmail.com for details.",
    
    "Send your resume to hr.team@company.co.in or careers@startup.ai",
    
    "My old email was dhuru_old@mail.com, new one is dhuru.new123@outlook.com",
    
    "For support: support@service.org! We are available 24/7.",
    
    "Emails: test.user+ml@gmail.com, admin_123@yahoo.co.uk",
    
    "Invalid emails like user@@gmail..com should not be matched.",
    
    "Reach us at: info@company.com; sales@company.com",
    
    "Personal: dhuru123@gmail.com | Work: dhuru@company.org",
    
    "Contact: <helpdesk@domain.com>",
    
    "Email me (dhuru.ml@ai-tech.io) for collaboration.",
    
    "Fake email: user#gmail.com (should not match)",
    
    "Multiple emails in one line: first@mail.com second@mail.org third123@domain.net"
]
for text in texts:
    email=re.findall(
        r"[a-zA-Z0-9._+]+@[a-zA-Z-]+\.[a-zA-Z.]+",
        text
    )
    print(f"'{text}' email : {email}")
    print()
```

### **Check if any digit exists**

```python
text = "No numbers here"
digits_check=re.search(r"\d",text)
digits_res="digits found" if digits_check else "No digits found"
print(digits_res)
```

```python
text = "Hello!!! This is ML 2026 😊"
cleaned_text=re.sub(r"[^a-zA-Z\s]","",text.lower())
print(f'Cleaned text : {cleaned_text}')
```

### **Extract #ML, #AI**

```python
text = "Learning #ML and #AI is fun"
hashtag_words=re.findall(r"#\w+",text)
print(f'Hashed words in the text : {hashtag_words}')
```

### **Extract: Dhuru, Machine, Learning**

```python
text = "Dhuru is learning Machine Learning"
specific_words=re.findall(r"\b[A-Z][a-zA-Z]*", text)
print(f'Specific words in the text : {specific_words}')
```

### **Remove the HTMLTags**

```python
text = "This is pandas and <p>it is great for data analysis</p>"
cleaned_text=re.sub(r"<.*?>","",text)
print(f'Text after removing HTML tags : {cleaned_text}')
```
